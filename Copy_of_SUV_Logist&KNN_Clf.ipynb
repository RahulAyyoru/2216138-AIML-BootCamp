```python
df=df.drop('User ID',axis=1)
df=df.drop('Gender',axis=1)
```

Data Preprocessing -Cleaning and organizing raw data

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1:].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
sst=StandardScaler()

x_train = sst.fit_transform(x_train)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
#predicting
y_pred = model.predict(sst.transform(x_test))
```

```python
print("Predicted value of training data:",model.score(x_train,y_train))
print("Predicted value of training data:",model.score(sst.transform(x_test),y_test))
```

```python
accuracy= model.score(x,y)*100
print(accuracy)
```

Deployment - Bringing our ML model into existence(action)

```python
age =int(input("Enter age:"))
est_sal= int(input("Enter estimated salary:"))
output=model.predict(sst.transform([[age,est_sal]]))
print(output)
if output[0]==1:
  print("The person will purchase an SUV")
else:
  print("The person wont purchase an SUV")
```
